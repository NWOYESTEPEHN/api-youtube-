```python
#import libraries
import requests
import pandas as pd
import time
```

```python
#keyaqR5g
API_KEY = "xxxxxxxxxxxx"
CHANNEL_ID = "MrBeast6000"
```

```python
#make API call
pageToken =""
url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelid="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=1000000"+pageToken

response = requests.get(url).json()
```

```python
response["items"][1]
```

```python
#build our dataframe
df = pd.DataFrame(columns=["video_id","video_title","upload_date","view_count","like_count","comment_count"])
```

```python
df
```

```python
for video in response['items']:
    if video['id']['kind'] == "youtube#video":
      video_id = video['id']['videoId']
      video_title = video['snippet']['title']
      video_title = str(video_title).replace("&","")
      upload_date = response['items'][0]['snippet']['publishedAt']
      upload_date = str(upload_date).split('T')[0]

      #collecting view
      url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
      response_video_stats = requests.get(url_video_stats).json()

      view_count = response_video_stats['items'][0]['statistics']['commentCount']
      comment_count = response_video_stats['items'][0]['statistics']['commentCount']

      #save data in pandas df
      df = df.append({'video_id':video_id,'video_title':video_title,
                      "upload_date":upload_date,
                      "like_count":like_count,
                      "comment_count":comment_count}, ignore_index=True)
```
